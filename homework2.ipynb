```python
import cv2  
import numpy as np  
import matplotlib.pyplot as plt
from math import cos, sin, radians
```

```python
img = cv2.imread('dataset/lena.jpg')
grayImage = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
height = grayImage.shape[0]
width = grayImage.shape[1]
```

#### 图像平移

```python
transform = np.array([[1, 0, 20], [0, 1, 20], [0, 0, 1]])
result = np.zeros((height, width))
for i in range(height):
    for j in range(width):
        original_pos = np.array([i, j, 1])
        target_pos = np.dot(transform, original_pos)
        x = int(target_pos[0])
        y = int(target_pos[1])
        if x >= height or y >= width or x < 0 or y < 0:
            result[i, j] = 255
        else:
            result[i, j] = grayImage[x, y]
plt.figure(figsize=(10, 12))
plt.subplot(121), plt.imshow(grayImage, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(result, 'gray'), plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```

#### 图像镜像

```python
transform = np.array([[1, 0, 0], [0, -1, height-1], [0, 0, 1]])
result = np.zeros((height, width), np.uint8)
for i in range(height):
    for j in range(width):
        original_pos = np.array([i, j, 1])
        target_pos = np.dot(transform, original_pos)
        x = int(target_pos[0])
        y = int(target_pos[1])
        if x >= height or y >= width or x < 0 or y < 0:
            result[i, j] = 255
        else:
            result[i, j] = grayImage[x, y]
plt.figure(figsize=(10, 12))
plt.subplot(121), plt.imshow(grayImage, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(result, 'gray'), plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```

#### 图像旋转

```python
beta = radians(15)
transform = np.array(
    [[cos(beta), -sin(beta), 0], [sin(beta), cos(beta), 0], [0, 0, 1]])
result = np.zeros((height, width), np.uint8)
for i in range(height):
    for j in range(width):
        original_pos = np.array([i, j, 1])
        target_pos = np.dot(transform, original_pos)
        x = int(target_pos[0])
        y = int(target_pos[1])
        if x >= height or y >= width or x < 0 or y < 0:
            result[i, j] = 255
        else:
            result[i, j] = grayImage[x, y]
plt.figure(figsize=(10, 12))
plt.subplot(121), plt.imshow(grayImage, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(result, 'gray'), plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```

#### 组合变换

```python
transform = np.array([[1, 0, 0], [0, -1, height-1], [0, 0, 1]])
result = np.zeros((height, width), np.uint8)
for i in range(height):
    for j in range(width):
        original_pos = np.array([i, j, 1])
        target_pos = np.dot(transform, original_pos)
        x = int(target_pos[0])
        y = int(target_pos[1])
        if x >= height or y >= width or x < 0 or y < 0:
            result[i, j] = 255
        else:
            result[i, j] = grayImage[x, y]
beta = radians(15)
transform = np.array(
    [[cos(beta), -sin(beta), 0], [sin(beta), cos(beta), 0], [0, 0, 1]])
result2 = np.zeros((height, width), np.uint8)
for i in range(height):
    for j in range(width):
        original_pos = np.array([i, j, 1])
        target_pos = np.dot(transform, original_pos)
        x = int(target_pos[0])
        y = int(target_pos[1])
        if x >= height or y >= width or x < 0 or y < 0:
            result2[i, j] = 255
        else:
            result2[i, j] = result[x, y]
plt.figure(figsize=(10, 12))
plt.subplot(121), plt.imshow(grayImage, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(result2, 'gray'), plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```
