```python
import cv2  
import numpy as np  
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('dataset/lena.jpg')
grayImage = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
height = grayImage.shape[0]
width = grayImage.shape[1]
```

### 线性运算

#### 线性平移

```python
result = np.zeros((height, width), np.uint8)
for i in range(height):
    for j in range(width):
        if (int(grayImage[i, j]+100) > 255):
            gray = 255
        else:
            gray = int(grayImage[i, j]+100)
        result[i, j] = np.uint8(gray)
plt.figure(figsize=(10, 12))
plt.subplot(121), plt.imshow(grayImage, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(result, 'gray'), plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```

#### 线性增强

```python
result = np.zeros((height, width), np.uint8)
for i in range(height):
    for j in range(width):
        if (int(grayImage[i,j]*2) > 255):
            gray = 255
        else:
            gray = int(grayImage[i,j]*2)
        result[i,j] = np.uint8(gray)
plt.figure(figsize=(10, 12))
plt.subplot(121), plt.imshow(grayImage, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(result, 'gray'), plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```

#### 线性减弱

```python
result = np.zeros((height, width), np.uint8)
for i in range(height):
    for j in range(width):
        gray = int(grayImage[i,j]*0.8)
        result[i,j] = np.uint8(gray)
plt.figure(figsize=(10, 12))
plt.subplot(121), plt.imshow(grayImage, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(result, 'gray'), plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```

#### 反色变换

```python
result = np.zeros((height, width), np.uint8)
for i in range(height):
    for j in range(width):
        gray = 255 - grayImage[i,j]
        result[i,j] = np.uint8(gray)
plt.figure(figsize=(10, 12))
plt.subplot(121), plt.imshow(grayImage, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(result, 'gray'), plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```

#### 分段线性运算

```python
result = np.zeros((height, width), np.uint8)
for x in range(1,height):
    for y in range(1,width):
        if(grayImage[x,y] <= 50):  
            result[x,y] = int(grayImage[x,y] *50 / 35.0)
        else:
            if(grayImage[x,y] > 50 and grayImage[x,y] <= 75):
                result[x,y] = int((150 / 40.0) * (grayImage[x,y] - 50) + 50)  
            else:
                result[x,y] = int((55 / 180.0) * (grayImage[x,y] - 75) + 200)
plt.figure(figsize=(10, 12))
plt.subplot(121), plt.imshow(grayImage, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(result, 'gray'), plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```

### 非线性点运算

#### 幂次变换

```python
result = np.zeros((height, width), np.uint8)
for i in range(height):
    for j in range(width):
        gray = int(grayImage[i,j]**0.5)
        result[i,j] = np.uint8(gray)
plt.figure(figsize=(10, 12))
plt.subplot(121), plt.imshow(grayImage, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(result, 'gray'), plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```

#### 对数变换

```python
def log(c, img):
    result = c * np.log(1.0 + img)
    return result
result = np.zeros((height, width), np.uint8)
for i in range(height):
    for j in range(width):
        result[i,j] = log(2,grayImage[i,j])
plt.figure(figsize=(10, 12))
plt.subplot(121), plt.imshow(grayImage, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(result, 'gray'), plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```
